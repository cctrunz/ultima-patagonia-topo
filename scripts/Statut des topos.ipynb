```python
INDEX_DIRECTORIES = [x for x in directories[:22] if (x != "../data/cadastre") and (x != "../data/ovni")]
```

```python
FIXES = []
for d in INDEX_DIRECTORIES:
    FILENAME = d+ '/'+d[-3:]+'.th'
    try: 
        with open(FILENAME, 'r') as f:
            for line in f.readlines():
                FIXES.append(line)
            f.close()
    except FileNotFoundError:
        pass
```

```python
TEMPLATE = '''
survey {survey} -title "{title}" -attr cad_num {cad_num} -attr exped UP{exped}

endsurvey
'''
```
