```python
import os
from os.path import isfile, join, dirname, abspath
import sys
import re
import argparse
import shutil
import subprocess
import tempfile
import pandas as pd
import pyproj as proj
```

```python
CAVITES = pd.read_csv('../data/cadastre/UP_MDD_DDA_Temp_synthese_Cavites.csv')
```

```python
CAVITES.columns
```

```python
cad_nums = {}
for dirpath, dirnames, filenames in os.walk('../data'):
    # check along each directory.
    for name in dirnames:
        # find all cave folders, each has a three digit cadastre sector identifier in path
        if not re.match('\d{3}', name):
            if re.match('.*\d{3}',dirpath) and (name !='outputs'):
                #print('yay!', name,dirpath)
                try:
                    print(CAVITES.loc[name]['NomCadastre'])
                except KeyError:
                    try:
                        
                        with open(dirpath+'/'+name+"/"+name+".th", "r") as f:
                            line = f.readlines()[1].split('-attr cad_num ')[1]
                            line = line.split(' -attr exped')[0]
                            print(line.split(" ")[0] )
                            f.close()
                            cad_nums = {**cad_nums, name:line.split(" ")[0]}
                    except FileNotFoundError:
                        print('no therion file for this cave yet?')
                        pass
                    #print("cette grotte n'est pas dans le xls")
```

```python
CAVITES['CAD_NUM'] = CAVITES['NomCadastre'].map(cad_nums)
```

```python
LATLONG = proj.CRS('epsg:4326')
UTM18S = proj.CRS('epsg:32718')
```

```python
transformer = proj.Transformer.from_crs("epsg:32718", "epsg:4326")
```

```python
transformer.transform(490088,4452354,105)
```

```python
# syntax for map_box. 

GEOJSON_TEMPLATE = """
var geojsonFeature = {{
"type": "FeatureCollection",
"name": "cave_sites",
"crs": {{ "type": "name", "properties": {{ "name": "urn:ogc:def:crs:OGC:1.3:CRS84" }} }},
"features": [{FEATURES}]
}};
"""
LINK = 'https://tr1813.github.io/ultima-patagonia-topo/html/'
FEATURE = '{{ "type": "Feature", "properties": {{ "id": {ID},"no_cadastral": "{no_cadastral}", "name": "{NomCadastre}", "developement": "{Developpement}", "depth": "{Denivellation}", "url": "url"}}, "geometry": {{ "type": "Point", "coordinates": [ {Y}, {X} ] }} }}'
```

```python
FEATURES = ""
for x,y,z,nomcadastre,ID,DevTopo,Prof,NUM_CAD in CAVITES[['X (UTM 18 Sud)','Y (UTM 18 Sud)','Alt.','NomCadastre','Unnamed: 0', 'Dev. Topo', 'Prof.','CAD_NUM']].values:
    try:
        x2,y2,z2 = transformer.transform(x,y,z)
        #print(nomcadastre)
        
        fmt_feature = FEATURE.format(X = x2,Y = y2, NomCadastre = nomcadastre, ID = ID, Developpement=DevTopo,Denivellation=Prof,no_cadastral=NUM_CAD)
        if 'inf' not in fmt_feature and '[ nan' not in fmt_feature:
            FEATURES+=fmt_feature+',\n'
    except TypeError:
        pass
        
print(FEATURES[:-3])
```

```python
print(GEOJSON_TEMPLATE.format(FEATURES = FEATURES[:-2]))
```

```python
with open('../../javascript/map_box_template.txt', 'r') as f:
    template = f.readlines()
    f.close()
    
```

```python
template.insert(66,GEOJSON_TEMPLATE.format(FEATURES = FEATURES[:-2]))
```
